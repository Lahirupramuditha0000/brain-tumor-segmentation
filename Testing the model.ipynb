```python
import joblib 
from matplotlib import pyplot as plt

model=joblib.load("brain_tumor_kmeans.sav")
```

```python
tumorlabel=4
```

```python
import cv2

imgpath="test_images/0(239).jpg"

img=cv2.imread(imgpath,0)

height,width=img.shape
image_flatten=img.reshape(height*width,1)

print("image shape: ",img.shape)
print("shape after flatten: ",image_flatten.shape)

plt.imshow(img,cmap="gray")
```

```python
labels=model.predict(image_flatten)

labels2D=labels.reshape(height,width)

plt.imshow(labels2D)
plt.colorbar()
plt.show()
```

```python
mask=(labels2D==tumorlabel)

plt.imshow(mask,cmap="gray")
plt.savefig("mask.png")
plt.colorbar()
plt.show()
```
